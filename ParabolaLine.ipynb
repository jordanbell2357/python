```python
from sympy import *
a = Symbol('a', real=True)
x = Symbol('x')
f = a*x**2
f
```

```python
xA = Symbol('x_A', real=True)
yA = Symbol('y_A', real=True)
yA = f.subs(x, xA)
A = Symbol('A')
A = Point2D(xA, yA)
A
```

```python
m = Symbol('m', real=True)
fm = Symbol('f_m')
fm = m*(x-xA) + yA
fm
```

```python
gm = Symbol('g_m')
gm = f-fm
gm
```

```python
Dm = Symbol('D_m')
Dm = discriminant(gm)
Dm
```

```python
factor(Dm)
```

```python
solve(factor(Dm), m)
```

```python
mA = Symbol('m_A')
mA = solve(factor(Dm), m)[0]
mA
```

```python
fA = Symbol('fA')
fA = fm.subs(m, mA)
fA
```

```python
expand(fA, x)
```
