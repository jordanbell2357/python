```python
from sympy import *
a = Symbol('a')
x = Symbol('x')
f = Symbol('f')
f = a*x**2
f
```

```python
xA = Symbol('x_A')
yA = Symbol('y_A')
yA = f.subs(x, xA)
A = Symbol('A')
A = Point2D(xA, yA)
A
```

```python
m = Symbol('m')
fm = Symbol('f_m')
fm = m*(x-xA) + yA
fm
```

```python
gm = Symbol('g_m')
gm = f-fm
gm
```

```python
Dm = Symbol('D_m')
Dm = discriminant(gm)
Dm
```

```python
factor(Dm)
```

```python
solve(Dm, m)
```

```python
mA = Symbol('m_A')
mA = solve(factor(Dm), m)[0]
mA
```

```python
fA = Symbol('fA')
fA = fm.subs(m, mA)
fA
```

```python
expand(fA, x)
```

```python
x0 = 2
a0 = Rational(1, 4)
p1 = plot(f.subs(a, a0), show=False)
p2 = plot(fA.subs([(a, a0), (xA, x0)]), show=False)
p1.extend(p2)
p1.show()
```
